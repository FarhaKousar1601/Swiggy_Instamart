```python
# Merge datasets on 'OrderID'
df = pd.merge(order_details, orders, on='OrderID')

# Convert datetime columns
df['OrderDateTime'] = pd.to_datetime(df['OrderDateTime'])
df['DeliveryDateTime'] = pd.to_datetime(df['DeliveryDateTime'])

# Add DeliveryTime in minutes
df['DeliveryTime(mins)'] = (df['DeliveryDateTime'] - df['OrderDateTime']).dt.total_seconds() / 60

# Calculate TotalProductPrice
df['TotalProductPrice'] = df['Quantity'] * df['PricePerUnit']

print("Basic Info:")
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
```

```python
print("\n--- KPIs ---")
print("Total Orders:", df['OrderID'].nunique())
print("Total Customers:", df['CustomerID'].nunique())
print("Total Revenue (from order_details): ₹", df['TotalProductPrice'].sum())
print("Avg Order Value: ₹", df.groupby('OrderID')['TotalProductPrice'].sum().mean())
print("Avg Delivery Time (mins):", df['DeliveryTime(mins)'].mean())
```

```python
top_products = df.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_products.values, y=top_products.index, palette="viridis")
plt.title("Top 10 Most Ordered Products")
plt.xlabel("Total Quantity Sold")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='DeliveryStatus', palette='pastel')
plt.title("Delivery Status Distribution")
plt.show()
```

```python
df['OrderDate'] = df['OrderDateTime'].dt.date
daily_sales = df.groupby('OrderDate')['TotalProductPrice'].sum()

plt.figure(figsize=(12,5))
daily_sales.plot(kind='line', marker='o')
plt.title("Daily Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue (₹)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['OrderHour'] = df['OrderDateTime'].dt.hour
hourly_orders = df.groupby('OrderHour')['OrderID'].count()

plt.figure(figsize=(10,4))
sns.lineplot(x=hourly_orders.index, y=hourly_orders.values, marker="o")
plt.title("Hourly Order Volume")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Orders")
plt.xticks(range(0,24))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['DayOfWeek'] = df['OrderDateTime'].dt.day_name()
heatmap_data = df.groupby(['DayOfWeek', 'OrderHour'])['OrderID'].count().unstack().fillna(0)

# Order the days of the week
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
heatmap_data = heatmap_data.reindex(ordered_days)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap='YlOrBr', linewidths=0.5)
plt.title("Heatmap of Orders by Day and Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Day of Week")
plt.tight_layout()
plt.show()
```
